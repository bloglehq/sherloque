```python
import nltk
nltk.download("punkt_tab")
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

stemmer = PorterStemmer()
```

```python
text = "How we leveraged vLLM to power our GenAI applications at LinkedIn"
```

```python
toks = word_tokenize(text)
toks
```

```python
for w in toks:
    print(stemmer.stem(w))
```

```python
l = WordNetLemmatizer()
```

```python
for w in toks:
    print(l.lemmatize(w))
```
