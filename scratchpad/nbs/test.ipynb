```python
from config import get_settings
import psycopg
from psycopg import sql
```

```python
url_ids = [1, 2, 3]
```

```python
from sherloque.utils import preprocess_text

query = "services and monolithic system design"
qtoks = await preprocess_text(query)
qtoks
```

```python
qtids = []
async with await psycopg.AsyncConnection.connect(get_settings().DB_URL) as conn:
    async with conn.cursor() as cursor:
        await cursor.execute(
            sql.SQL("SELECT _id FROM token_list WHERE token IN ({})").format(
                sql.SQL(", ").join([sql.Literal(token) for token in qtoks])
            )
        )
        qtids = list(map(lambda x: x[0], await cursor.fetchall()))

qtids
```

```python
async with await psycopg.AsyncConnection.connect(get_settings().DB_URL) as conn:
    async with conn.cursor() as cursor:
        await cursor.execute(
            sql.SQL("""
            SELECT url_id, COUNT(token_id)
            FROM token_location
            WHERE token_id IN ({}) AND url_id IN ({})
            GROUP BY url_id
            """).format(
                sql.SQL(", ").join([sql.Literal(qtid) for qtid in qtids]),
                sql.SQL(", ").join([sql.Literal(url_id) for url_id in url_ids])
            )
        )
        print(await cursor.fetchall())
```

```python
async with await psycopg.AsyncConnection.connect(get_settings().DB_URL) as conn:
    async with conn.cursor() as cursor:
        await cursor.execute(
            sql.SQL("SELECT url_id, token_id FROM token_location WHERE url_id IN ({})").format(
                sql.SQL(", ").join([sql.Literal(url_id) for url_id in url_ids])
            )
        )
        print(await cursor.fetchall())
```
