```python
from sherloque.retrieve import BM25Retriever, VectorRetriever, BM25RetrieverConfig, VectorRetrieverConfig
from sherloque.rank import CrossEncoderReRanker
from sherloque.query_engine import QueryEngine
from sherloque.config import get_async_engine
from dotenv import load_dotenv

load_dotenv("/Users/suchitg/gen/sherloque/src/sherloque/.env")

engine = get_async_engine()
```

```python
bc = BM25RetrieverConfig(
    top_k=10
)

vc = VectorRetrieverConfig(
    embedding_model="nomic-ai/nomic-embed-text-v1.5",
    top_k=10,
)
```

```python
b = BM25Retriever(engine, config=bc)
v = VectorRetriever(engine, config=vc)
r = CrossEncoderReRanker(engine)
```

```python
bres = await b.retrieve(query="T cells immune response")
print(bres)
len(bres)
```

```python
vres = await v.retrieve(query="T cells immune response")
print(vres)
len(vres)
```

```python
qe = QueryEngine(retrievers=[b, v], rankers=[r])
res = await qe.search(query="T cells immune response")
print(res)
```
